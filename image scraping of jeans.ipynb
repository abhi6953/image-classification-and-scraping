```python
search_bar = driver.find_element_by_id("twotabsearchtextbox")   
search_bar.clear()                                               

input_category = input('name of category you want to search: ')
search_bar.send_keys(input_category)
```

```python
#search the images
button = driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]/span/input')     
button.click()

#find image tag
images = driver.find_elements_by_xpath('//img[@class="s-image"]')

#importing the lenghts
img_urls = []
img_data = []
for page in range(0,3):
    for image in images:
        source= image.get_attribute('src')
        if source is not None:
            if(source[0:4] == 'http'):
                img_urls.append(source)
len(img_urls)

#downloading the images
for i in range(len(img_urls)):
    if i >= len(img_urls):
        break
    print("Downloading {0} of {1} images" .format(i, len(img_urls)))
    response= requests.get(img_urls[i])
    file = open(r"D:\New folder\DAta science\Image Scraping and Classification Project\img\jeans"+str(i)+".jpg", "wb")
    file.write(response.content)
```
